```python
import cv2
import matplotlib.pyplot as plt
from skimage import io, feature
```

```python
#load an example image (you can replace this with your own image)
image = io.imread('test.jpg')

#define hog parameters
cell_size = (8,8) #size of each cell in pixels
block_size = (2,2) #size of each block in cells
bins= 9 #number of orientation bins
```

```python
#calculate HOG features with block normalization
hog_features = feature.hog(image, pixels_per_cell= cell_size, cells_per_block=block_size,visualize=False, block_norm='L2-Hys', feature_vector= True, multichannel= True, orientations=bins)
#grayscale_img.jpg was not taken
```

```python
#Visualize the HOG image(optinal)
_, hog_image = feature.hog(image, pixels_per_cell= cell_size, cells_per_block=block_size, visualize=True, block_norm='L2-Hys',multichannel=True, orientations=bins)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(hog_image, cmap=plt.cm.gray)
plt.axis('off')
plt.title('HOG Feature Visualization')
plt.show()

'''the hog_features variable now contains the HOG features with block normalization'''
```
